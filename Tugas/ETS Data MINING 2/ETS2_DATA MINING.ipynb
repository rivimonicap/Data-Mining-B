```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv('online_retail.csv')
data.head()
```

```python
data['Description'] = data['Description'].str.strip()
```

```python
data.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
```

```python
data['InvoiceNo'] = data['InvoiceNo'].astype('str')
```

```python
data = data[~data['InvoiceNo'].str.contains('C')]
```

```python
data2 = (data[data['Country'] =="France"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
data2.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
data2_sets = data2.applymap(encode_units)
```

```python
data2_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
data2_sets.head()
```

```python
data3 = apriori(data2_sets, min_support=0.06, use_colnames=True)
data3.head()
```

```python
data4 = association_rules(data3, metric="lift", min_threshold=1)
data4.head()
```

```python
data4[ (data4['lift'] >= 8) &
       (data4['confidence'] >= 0.6) ]
```
