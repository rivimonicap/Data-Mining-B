```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
data = pd.read_csv('online_retail.csv')
data.head()
```

```python
data.info()
```

```python
np.sum(data.isnull())
```

```python
data['Description'] = data['Description'].str.strip()
```

```python
data.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
data['InvoiceNo'] = data['InvoiceNo'].astype('str')
data = data[~data['InvoiceNo'].str.contains('C')]
```

```python
data['UnitPrice'] = data['UnitPrice'].astype('str')
data = data[~data['UnitPrice'].str.contains('0')]
```

```python
np.sum(data.isnull())
```

```python
g = data.groupby('InvoiceNo')
data2 = []
for i,j in g:
    data2.append(list(j['Description'].map(str)))
```

```python
data2 = pd.DataFrame(data2)
data2
```

```python
data2.shape
```

```python
records=[]
for i in range (0,19168) :
    records.append([str(data2.values[i,j]) for  j in range (0, 923)])
```

```python
arules = apriori(records, min_support=0.04,
            min_confidence=0.4,
            min_lift=3, min_length=2)
aresult=list(arules)
```

```python
aresult[0]
```

```python
for item in aresult:
    pair = item[0]
    items = [x for x in pair]
    print ("rule: " + items[0] + "->" + items[1])
    
    print ("support: " + str(item[1]))
    
    print ("confidence: " + str(item[2][0][2]))
    
    print ("lift: " + str(item[2][0][3]))
    
    print ("============================================")
```
