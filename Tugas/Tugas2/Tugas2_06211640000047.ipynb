## Tugas2 Preprocessing

```python
data['MinTemp'] = data['MinTemp'].fillna((data['MinTemp'].mean())) 
data['MaxTemp'] = data['MaxTemp'].fillna((data['MaxTemp'].median()))
data['Rainfall'] = data['Rainfall'].fillna((data['Rainfall'].mean())) 
data['Evaporation'] = data['Evaporation'].fillna((data['Evaporation'].median()))
data['Sunshine'] = data['Sunshine'].fillna((data['Sunshine'].mean())) 
data['WindGustSpeed'] = data['WindGustSpeed'].fillna((data['WindGustSpeed'].median()))
data['WindSpeed9am'] = data['WindSpeed9am'].fillna((data['WindSpeed9am'].mean())) 
data['WindSpeed3pm'] = data['WindSpeed3pm'].fillna((data['WindSpeed3pm'].median()))
data['Humidity9am'] = data['Humidity9am'].fillna((data['Humidity9am'].mean())) 
data['Humidity3pm'] = data['Humidity3pm'].fillna((data['Humidity3pm'].median()))
data.isnull().any()
```

```python
data['Pressure9am'] = data['Pressure9am'].fillna(99) 
data['Pressure3pm'] = data['Pressure3pm'].fillna(111)
data['Cloud9am'] = data['Cloud9am'].fillna(99) 
data['Cloud3pm'] = data['Cloud3pm'].fillna(111)
data['Temp9am'] = data['Temp9am'].fillna(999) 
data['Temp3pm'] = data['Temp3pm'].fillna(1111)
data.isnull().any()
```

```python
data['RainTomorrow'] = data.fillna(data['RainTomorrow'].value_counts().index[1])
data.isnull().any()
```

```python
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

## Encode Label

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['MinTemp'] = LabelEncoder().fit_transform(data['MinTemp'])
data['MaxTemp'] = LabelEncoder().fit_transform(data['MaxTemp'])
data['Rainfall'] = LabelEncoder().fit_transform(data['Rainfall'])
data['Evaporation'] = LabelEncoder().fit_transform(data['Evaporation'])
data['Sunshine'] = LabelEncoder().fit_transform(data['Sunshine'])
data['WindGustDir'] = LabelEncoder().fit_transform(data['WindGustDir'])
data['WindGustSpeed'] = LabelEncoder().fit_transform(data['WindGustSpeed'])
data['WindDir9am'] = LabelEncoder().fit_transform(data['WindDir9am'])
data['WindDir3pm'] = LabelEncoder().fit_transform(data['WindDir3pm'])
data['WindSpeed9am'] = LabelEncoder().fit_transform(data['WindSpeed9am'])
data['WindSpeed3pm'] = LabelEncoder().fit_transform(data['WindSpeed3pm'])
data['Humidity9am'] = LabelEncoder().fit_transform(data['Humidity9am'])
data['Humidity3pm'] = LabelEncoder().fit_transform(data['Humidity3pm'])
data['Pressure9am'] = LabelEncoder().fit_transform(data['Pressure9am'])
data['Pressure3pm'] = LabelEncoder().fit_transform(data['Pressure3pm'])
data['Cloud9am'] = LabelEncoder().fit_transform(data['Cloud9am'])
data['Cloud3pm'] = LabelEncoder().fit_transform(data['Cloud3pm'])
data['Temp9am'] = LabelEncoder().fit_transform(data['Temp9am'])
data['Temp3pm'] = LabelEncoder().fit_transform(data['Temp3pm'])
data['RainToday'] = LabelEncoder().fit_transform(data['RainToday'])
data['RISK_MM'] = LabelEncoder().fit_transform(data['RISK_MM'])
data['RainTomorrow'] = LabelEncoder().fit_transform(data['RainTomorrow'])
data.head()
```
